Patient registration:

5. Push triage assessment(Health Complain, Temperature, Heartbeat, Blood Pressure) information to patient file (H2) use as stack structure. (average run time= O(1))
6. As per the triage level, to build priority queue(heap sort-min heap). (average run time= O(log n)
7. use queue structure to dequeue the patient (average run time= O(1)), provide est wait time and queue #

2. update the queue at priority queue by min heap.


```python
class Max_Min_BinaryHeap:
    def __init__(self):
        self.heap = []

    def Left(self, i):
        l_child_index = 2 * i + 1
        return l_child_index

    def Right(self, i):
        r_child_index = 2 * i + 2
        return r_child_index

    def Parent(self, i):
        parent_node_index = (i - 1) // 2
        return parent_node_index

    def insert(self, element, is_max_heap=True):
        self.heap.append(element)
        i = len(self.heap) - 1
        if is_max_heap:
            self.shiftup_Max(i)
        else:
            self.shiftup_Min(i)

    def delete(self, element, is_max_heap=True):
        if element in self.heap:
            i=self.heap.index(element)
            self.heap[i],self.heap[-1]=self.heap[-1],self.heap[i]
            self.heap.pop()
            if True:
                self.shiftdown_Max_Heapify(i,len(self.heap))
            else:
                self.shiftdown_Min_Heapify(i,len(self.heap))
    def shiftup_Max(self, i):
        while i > 0 and self.heap[i] > self.heap[self.Parent(i)]:
            self.heap[i], self.heap[self.Parent(i)] = self.heap[self.Parent(i)], self.heap[i]
            i = self.Parent(i)

    def shiftup_Min(self, i):
        while i > 0 and self.heap[i] < self.heap[self.Parent(i)]:
            self.heap[i], self.heap[self.Parent(i)] = self.heap[self.Parent(i)], self.heap[i]
            i = self.Parent(i)

    def shiftdown_Max_Heapify(self, i, n):
        while self.Left(i) < n:
            largest = self.Left(i)
            if self.Right(i) < n and self.heap[self.Right(i)] > self.heap[self.Left(i)]:
                largest = self.Right(i)
            if self.heap[i] < self.heap[largest]:
                self.heap[i], self.heap[largest] = self.heap[largest], self.heap[i]
                i = largest
                self.shiftdown_Min_Heapify(i, n)
            else:
                break

    def shiftdown_Min_Heapify(self, i, n):
        while self.Left(i) < n:
            smallest = self.Left(i)
            if self.Right(i) < n and self.heap[self.Right(i)] < self.heap[smallest]:
                smallest = self.Right(i)
            if self.heap[i] > self.heap[smallest]:
                self.heap[i], self.heap[smallest] = self.heap[smallest], self.heap[i]
                i = smallest
            else:
                break

    def Build_heap(self, Arr, is_max_heap=True):
        self.heap = Arr[:]
        n = len(self.heap)
        if is_max_heap:
            for i in range(n // 2 - 1, -1, -1):
                self.shiftdown_Max_Heapify(i, n)
        else:
            for i in range(n // 2 - 1, -1, -1):
                self.shiftdown_Min_Heapify(i, n)

    def HeapSort(self, Arr, is_max_heap=True):
        self.heap = Arr[:]
        self.Build_heap(Arr, is_max_heap=is_max_heap)
        sorted_heap = []

        n = len(Arr)
        for i in range(n - 1, 0, -1):
            self.heap[0], self.heap[i] = self.heap[i], self.heap[0]

            if is_max_heap:
                self.shiftdown_Max_Heapify(0, i)

            else:
                self.shiftdown_Min_Heapify(0, i)



        return self.heap


heap = Max_Min_BinaryHeap()
string_arr = 'DATASTRUCTURE'
for char in string_arr:
    heap.insert(char, is_max_heap=True)
    heap.delete('A',is_max_heap=True)

sorted_heap = heap.HeapSort(heap.heap, is_max_heap=True)
print(sorted_heap)

```
