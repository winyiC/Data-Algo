
Once the patient registration is completed, it needs to evaluate patient triage level.<br>
This binary tree is using level-order travesal/and Breath-first travesal to direct to the final assessment level.<br>
It looks at level by level. To obtain the result of the triage assessment, it needs to look at each level.<br>
If the input is "no", it directs to the node.right.<br>
If the input is "yes", it directs to the node.left.<br>
When the travesal reached the node.left and node.right=None, it will return the last_node.<br>
Tree:


                        Walking?
                       No/      \Yes
                Breathing?    Minimal
              No /       \Yes
               Dead     Respiration >30?
                         No/      \Yes
                       Immediate  Radial pulse present?
                                 No/      \Yes
                              Immediate    Patient Responsive?
                                           No/          \Yes
                                       Immediate       Delayed


The worst runtime is the height of the tree =O(log n), The best runtime is O(1)


```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# Build the binary tree
class Triage_tree:
    #Create tree node
    class TreeNode:
        def __init__(self, value):
            self.value = value
            self.left = None
            self.right = None

    def __init__(self):
        self.last_node = None
        self.path = []
        self.input_path = []
        self.root = self.TreeNode("Walking?")
#Create triage assessment tree
    def build_triage_tree(self):

        breathing = self.TreeNode("Breathing?")
        minimal = self.TreeNode("Minimal")
        respiration = self.TreeNode("Respiration > 30?")
        immediate = self.TreeNode("Immediate")
        radial_pulse = self.TreeNode("Radial pulse present?")
        patient_responsive = self.TreeNode("Patient Responsive?")
        delayed = self.TreeNode("Delayed")
        dead = self.TreeNode("Dead")

        self.root.left = breathing
        self.root.right = minimal
        breathing.left =dead
        breathing.right = respiration
        respiration.left = immediate
        respiration.right = radial_pulse
        radial_pulse.left = immediate
        radial_pulse.right = patient_responsive
        patient_responsive.left = immediate
        patient_responsive.right = delayed


        return self.root

 # Using the concept of Breath-first search
    def traverse_level(self, nodes):
        next_level_nodes = []
        for node in nodes:
            self.path.append(node.value)
            if node.left or node.right:
                user_input = input(f"{node.value}\nEnter 'Yes' or 'No': ")
                self.input_path.append(user_input.lower())

                if user_input.lower() == 'yes':
                    if node.right:
                        next_level_nodes.append(node.right)
                    else:
                        print(f"Reached a leaf node: {node.value}")
                        self.last_node=node.value

                elif user_input.lower() == 'no':
                    if node.left:
                        next_level_nodes.append(node.left)
                    else:
                        print(f"Reached a leaf node: {node.value}")
                        self.last_node=node.value

                else:
                    print("Invalid input. Please enter 'Yes' or 'No'.")

            else:
                print(f"Reached a leaf node: {node.value}")
                self.last_node=node.value

#the start to iterate from the next node
        if next_level_nodes:
            self.traverse_level(next_level_nodes)

    def traverse_binary_tree_level_order(self):
        self.path = []
        self.input_path = []
        self.traverse_level([self.root])

    def get_path(self):
        return self.path

    def get_input_path(self):
        return self.input_path

    def get_last_node(self):
        return self.last_node

    def triage_level(self):
        if self.get_last_node()=="Immediate":
            print(f"Priority 1\n")
            return 1
        elif self.get_last_node()=="Delayed":
            print(f"Priority 2\n")
            return 2
        elif self.get_last_node()=="Minimal":
            print(f"Priority 3\n")
            return 3
        elif self.get_last_node()=="Dead":
            print('Patient Dead\n')
            return 0
# Create the triage tree
def assessment():
    triage_tree = Triage_tree()
    triage_tree.build_triage_tree()

# Start traversing the binary tree level by level from the root node
    triage_tree.traverse_binary_tree_level_order()

# Get the path followed during traversal
    path_followed = triage_tree.get_path()
    input_given = triage_tree.get_input_path()

# Print path and input side by side
    print("{:<25} {:<10}".format("Path Followed", "Input Given"))
    for path, input_val in zip(path_followed, input_given):
        print("{:<25} {:<10}".format(path, input_val))

# Obtain Priority level
    assessment_result=triage_tree.triage_level()
    return assessment_result

```


Build a priority queue by referencing the assessment result. (average run time= O(log n)<br>
  * Max-Heap to sort the queue.<br>
  * Calculate the est wait time based on the queue.<br>
  * Once the patient is assigned a bed, use queue structure to dequeue the patient (average run time= O(1))
The drawback of this priority queue:
  * Due to it having limited space for the priority queue, it is not able to insert patients to the priority queue when the capacity is full. <br>
Therefore, we can use the stack data structure(FILO) to pop the last patient out of the queue and store this patient to the wait-list queue(FIFO). <br>



```python
class Max_Min_BinaryHeap:
    def __init__(self):
        self.heap = []
        self.patient_info = {}
        self.counter = 0
        self.stack = []
    def Left(self, i):
        l_child_index = 2 * i + 1
        return l_child_index

    def Right(self, i):
        r_child_index = 2 * i + 2
        return r_child_index

    def Parent(self, i):
        parent_node_index = (i - 1) // 2
        return parent_node_index
#insert element to the heap.
#The drawback of this heap sort: when it reaches to full capacity=300, it is not able to insert the item to the heap.
#To solve this issue, it can apply stack data structure concept to this.
#When len(self.heap)>300,
    #pop self.heap[-1]
#Then, the priority Patient can insert to the priority queue

    def insert(self, element):
        if len(self.heap)>300:
            return print('Full Capacity')
        else:
            if element['Triage_Level'] == 0:
                print('Patient Dead! Inform patient family')
            else:
                key =  self.counter
                self.counter += 1
                self.heap.append((key, element))
                i = len(self.heap) - 1
                self.shiftup_Max(i, element)

    def delete(self):
        if len(self.heap)==0:
            return print('Empty queue')
        else:
            root=self.heap[0]
            print(root)
            patient_data=self.patient_info.get(0)
            peek_element=self.heap.pop()
            self.patient_info.pop(0)
            if len(self.heap)>0:
                self.heap[0]=peek_element
                self.shiftdown_Max_Heapify(0,len(self.heap))
            return (root,patient_data)

    def peek(self):
        if not self.heap:
            return None
        item=self.patient_info.get(0)
        #print(item)
        return item

    def shiftup_Max(self, i, data):
        while i > 0 and self.heap[i][0] > self.heap[self.Parent(i)][0]:
            self.heap[i], self.heap[self.Parent(i)] = self.heap[self.Parent(i)], self.heap[i]
            i = self.Parent(i)
            self.patient_info[i] = data
        return self.patient_info

    def shiftdown_Max_Heapify(self, i, n):
        while self.Left(i) < n:
            largest = self.Left(i)
            if self.Right(i) < n and self.heap[self.Right(i)][0] > self.heap[self.Left(i)][0]:
                largest = self.Right(i)
            if self.heap[i][0] < self.heap[largest][0]:
                self.heap[i], self.heap[largest] = self.heap[largest], self.heap[i]
                i = largest
            else:
                break

    def Build_heap(self):
        n = len(self.heap)
        for i in range(n // 2 - 1, -1, -1):
            self.shiftdown_Max_Heapify(i, n)

    def HeapSort(self):

        self.Build_heap()

        n = len(self.heap)
        for i in range(n - 1, 0, -1):
            self.heap[0], self.heap[i] = self.heap[i], self.heap[0]
            self.shiftdown_Max_Heapify(0, i)

        return self.heap
# Example
print("Example for Binary Tree search and Heap Sort\n")
heap = Max_Min_BinaryHeap()

T1 = assessment()
P1 = {"Name": "Apple", "Triage_Level": T1}
T2 = assessment()
P2 = {"Name": "Capple", "Triage_Level": T2}
T3 = assessment()
P3 = {"Name": "Bapple", "Triage_Level": T3}

heap.insert(P1)
heap.insert(P2)
heap.insert(P3)
print(f'first heap sort{heap.HeapSort()}\n')
heap.delete()
print(f'after delete{heap.HeapSort()}\n')
T4 = assessment()
P4 = {"Name": "Bapple", "Triage_Level": T4}
print(f'after insert{heap.HeapSort()}\n')

print(f'End Examples')
```
