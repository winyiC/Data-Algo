Once the patient registration is completed,it needs to evaluate patient triage level.
This binary tree is using level-order travesal/and Breath-first travesal to direct to the final assessment level.
It looks at level by level. To obtain the result of triage assessment, it needs to look at each level.
If input is "no", it directs to node.right.
If input is "yes", it directs to node.left.
When the travesal reached to the node.left and node.right=None, it will return the last_node
Tree:
 Level0                        Walking?
 Level1                       No/      \Yes
 Level2                Breathing ?    Minimal
 Level3              No /       \Yes
 Level4              Dead     Respiration >30?
 Level5                         No/      \Yes
 Level6                     Immediate  Radial pulse present?
 Level7                                 No/      \Yes
 Level8                             Immediate    Patient Responsive?
 Level9                                          No/          \Yes
 Level10                                       Immediate       Delayed

The worst runtime is the height of the tree =O(log n),
The best runtime is O(1)

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# Build the binary tree
class Triage_tree:
    #Create tree node
    class TreeNode:
        def __init__(self, value):
            self.value = value
            self.left = None
            self.right = None

    def __init__(self):
        self.last_node = None
        self.path = []
        self.input_path = []
        self.root = self.TreeNode("Walking?")
#Create triage assessment tree
    def build_triage_tree(self):

        breathing = self.TreeNode("Breathing?")
        minimal = self.TreeNode("Minimal")
        respiration = self.TreeNode("Respiration > 30?")
        immediate = self.TreeNode("Immediate")
        radial_pulse = self.TreeNode("Radial pulse present?")
        patient_responsive = self.TreeNode("Patient Responsive?")
        delayed = self.TreeNode("Delayed")
        dead = self.TreeNode("Dead")

        self.root.left = breathing
        self.root.right = minimal
        breathing.left =dead
        breathing.right = respiration
        respiration.left = immediate
        respiration.right = radial_pulse
        radial_pulse.left = immediate
        radial_pulse.right = patient_responsive
        patient_responsive.left = immediate
        patient_responsive.right = delayed


        return self.root

 # Using the concept of Breath-first search
    def traverse_level(self, nodes):
        next_level_nodes = []
        for node in nodes:
            self.path.append(node.value)
            if node.left or node.right:
                user_input = input(f"{node.value}\nEnter 'Yes' or 'No': ")
                self.input_path.append(user_input.lower())

                if user_input.lower() == 'yes':
                    if node.right:
                        next_level_nodes.append(node.right)
                    else:
                        print(f"Reached a leaf node: {node.value}")
                        self.last_node=node.value

                elif user_input.lower() == 'no':
                    if node.left:
                        next_level_nodes.append(node.left)
                    else:
                        print(f"Reached a leaf node: {node.value}")
                        self.last_node=node.value

                else:
                    print("Invalid input. Please enter 'Yes' or 'No'.")

            else:
                print(f"Reached a leaf node: {node.value}")
                self.last_node=node.value

#the start to iterate from the next node
        if next_level_nodes:
            self.traverse_level(next_level_nodes)

    def traverse_binary_tree_level_order(self):
        self.path = []
        self.input_path = []
        self.traverse_level([self.root])

    def get_path(self):
        return self.path

    def get_input_path(self):
        return self.input_path

    def get_last_node(self):
        return self.last_node

    def triage_level(self):
        if self.get_last_node()=="Immediate":
            print("Priority 1")
            return 1
        elif self.get_last_node()=="Delayed":
            print("Priority 2")
            return 2
        elif self.get_last_node()=="Minimal":
            print("Priority 3")
            return 3
        elif self.get_last_node()=="Dead":
            print('Patient Dead')
            return 0
# Create the triage tree
def assessment():
    triage_tree = Triage_tree()
    triage_tree.build_triage_tree()

# Start traversing the binary tree level by level from the root node
    triage_tree.traverse_binary_tree_level_order()

# Get the path followed during traversal
    path_followed = triage_tree.get_path()
    input_given = triage_tree.get_input_path()

# Print path and input side by side
    print("{:<25} {:<10}".format("Path Followed", "Input Given"))
    for path, input_val in zip(path_followed, input_given):
        print("{:<25} {:<10}".format(path, input_val))

# Obtain Priority level
    assessment_result=triage_tree.triage_level()
    return assessment_result

```

```python
class Max_Min_BinaryHeap:
    def __init__(self):
        self.heap = []

    def Left(self, i):
        l_child_index = 2 * i + 1
        return l_child_index

    def Right(self, i):
        r_child_index = 2 * i + 2
        return r_child_index

    def Parent(self, i):
        parent_node_index = (i - 1) // 2
        return parent_node_index

    def insert(self, element, is_max_heap=True):
        self.heap.append(element)
        i = len(self.heap) - 1
        if is_max_heap:
            self.shiftup_Max(i)
        else:
            self.shiftup_Min(i)

    def delete(self, element, is_max_heap=True):
        if element in self.heap:
            i=self.heap.index(element)
            self.heap[i],self.heap[-1]=self.heap[-1],self.heap[i]
            self.heap.pop()
            if True:
                self.shiftdown_Max_Heapify(i,len(self.heap))
            else:
                self.shiftdown_Min_Heapify(i,len(self.heap))
    def shiftup_Max(self, i):
        while i > 0 and self.heap[i] > self.heap[self.Parent(i)]:
            self.heap[i], self.heap[self.Parent(i)] = self.heap[self.Parent(i)], self.heap[i]
            i = self.Parent(i)

    def shiftup_Min(self, i):
        while i > 0 and self.heap[i] < self.heap[self.Parent(i)]:
            self.heap[i], self.heap[self.Parent(i)] = self.heap[self.Parent(i)], self.heap[i]
            i = self.Parent(i)

    def shiftdown_Max_Heapify(self, i, n):
        while self.Left(i) < n:
            largest = self.Left(i)
            if self.Right(i) < n and self.heap[self.Right(i)] > self.heap[self.Left(i)]:
                largest = self.Right(i)
            if self.heap[i] < self.heap[largest]:
                self.heap[i], self.heap[largest] = self.heap[largest], self.heap[i]
                i = largest
                self.shiftdown_Min_Heapify(i, n)
            else:
                break

    def shiftdown_Min_Heapify(self, i, n):
        while self.Left(i) < n:
            smallest = self.Left(i)
            if self.Right(i) < n and self.heap[self.Right(i)] < self.heap[smallest]:
                smallest = self.Right(i)
            if self.heap[i] > self.heap[smallest]:
                self.heap[i], self.heap[smallest] = self.heap[smallest], self.heap[i]
                i = smallest
            else:
                break

    def Build_heap(self, Arr, is_max_heap=True):
        self.heap = Arr[:]
        n = len(self.heap)
        if is_max_heap:
            for i in range(n // 2 - 1, -1, -1):
                self.shiftdown_Max_Heapify(i, n)
        else:
            for i in range(n // 2 - 1, -1, -1):
                self.shiftdown_Min_Heapify(i, n)

    def HeapSort(self, Arr, is_max_heap=True):
        self.heap = Arr[:]
        self.Build_heap(Arr, is_max_heap=is_max_heap)
        sorted_heap = []

        n = len(Arr)
        for i in range(n - 1, 0, -1):
            self.heap[0], self.heap[i] = self.heap[i], self.heap[0]

            if is_max_heap:
                self.shiftdown_Max_Heapify(0, i)

            else:
                self.shiftdown_Min_Heapify(0, i)



        return self.heap


heap = Max_Min_BinaryHeap()

T1=assessment()

Patient1={"Name":"Apple", "Triage_Level":T1}
T2=assessment()
P2={"Name":"Capple", "Triage_Level":T2}

heap.insert(Patient1['Triage_Level'], is_max_heap=True)
heap.insert(P2['Triage_Level'], is_max_heap=True)

sorted_heap = heap.HeapSort(heap.heap, is_max_heap=True)
print(sorted_heap)

```
