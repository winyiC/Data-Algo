```python


import numpy as np
import matplotlib.pyplot as plt



'%%time'
def insertion_run(arr):

  for j in range(1, len(arr)):
      key=arr[j]

      i=j-1

      while i>=0 and arr[i]>key:
          arr[i+1]=arr[i]

          i=i-1

      arr[i+1]=key

  return arr
arr=[4,5,6,7,0,4,8]
new_arr=insertion_run(arr)
print(new_arr)

```

```python


def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_arr = arr[:mid]
        right_arr = arr[mid:]
        merge_sort(left_arr)
        merge_sort(right_arr)
        i = j = k = 0
        while i < len(left_arr) and j < len(right_arr):
            if left_arr[i] <= right_arr[j]:
                arr[k] = left_arr[i]
                i += 1
            else:
                arr[k] = right_arr[j]
                j += 1
            k += 1
        while i < len(left_arr):
            arr[k] = left_arr[i]
            i += 1
            k += 1
        while j < len(right_arr):
            arr[k] = right_arr[j]
            j += 1
            k += 1

# Example usage
arr = [7, 5,4,9, 0, 3, 3, 1, 5, 0]
merge_sort(arr)
print(arr)
```
